```python
# ==========================================================
# Early Detection & Severity Classification of DR (PyTorch)
# ==========================================================
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as T

from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve, accuracy_score
from sklearn.utils.class_weight import compute_class_weight

# ----------------------------------------------------------
# GPU CHECK
# ----------------------------------------------------------
print("PyTorch:", torch.__version__)
print("CUDA Available:", torch.cuda.is_available())
gpu_count = torch.cuda.device_count()
print("Number of devices:", gpu_count)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ----------------------------------------------------------
# PATHS & CONFIG
# ----------------------------------------------------------
BASE_DIR = r"./"

CSV_PATH = os.path.join(BASE_DIR, "train.csv")
IMG_DIR = os.path.join(BASE_DIR, "train_images")

IMG_SIZE = 299
SEED = 42

# Setăm seed-urile pentru reproductibilitate
np.random.seed(SEED)
torch.manual_seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# ----------------------------------------------------------
# LOAD DATA
# ----------------------------------------------------------
df = pd.read_csv(CSV_PATH)
df["path"] = df["id_code"].apply(lambda x: os.path.join(IMG_DIR, f"{x}.png"))

print(df.head())
print(df["diagnosis"].value_counts())
print(df["path"].head())

# ----------------------------------------------------------
# PREPROCESSING
# ----------------------------------------------------------
def crop_fundus(img):
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    _, thresh = cv2.threshold(gray, 10, 255, cv2.THRESH_BINARY)
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if len(contours) > 0:
        c = max(contours, key=cv2.contourArea)
        x, y, w, h = cv2.boundingRect(c)
        img = img[y:y+h, x:x+w]

    return img

def preprocess_image(path):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # crop black borders
    img = crop_fundus(img)

    # resize
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))

    # CLAHE
    lab = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)
    l, a, b = cv2.split(lab)

    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    l = clahe.apply(l)

    lab = cv2.merge((l, a, b))
    img = cv2.cvtColor(lab, cv2.COLOR_LAB2RGB)

    # Gaussian blur
    img = cv2.GaussianBlur(img, (3,3), 0)

    # normalize
    img = img.astype("float32") / 255.0
    
    # PyTorch se așteaptă la ordinea canalelor (Canale, Înălțime, Lățime)
    img = np.transpose(img, (2, 0, 1))
    return img

# ----------------------------------------------------------
# LOAD IMAGES INTO MEMORY
# ----------------------------------------------------------
# Alocăm matricea numpy (N, C, H, W)
X = np.zeros((len(df), 3, IMG_SIZE, IMG_SIZE), dtype=np.float32)
y = df["diagnosis"].values

for i, path in enumerate(tqdm(df["path"])):
    X[i] = preprocess_image(path)

# ----------------------------------------------------------
# LABELS
# ----------------------------------------------------------
# Binary: 0 vs DR
y_binary = np.where(y == 0, 0, 1)

# Early Detection: 0=No DR, 1=Early(1,2), 2=Severe(3,4)
def map_early(v):
    if v == 0:
        return 0
    elif v in [1,2]:
        return 1
    else:
        return 2

y_early = np.array([map_early(v) for v in y])

# Severity: Original 5 classes
y_severity = y.copy()

# ----------------------------------------------------------
# DATASET & AUGMENTATION
# ----------------------------------------------------------
# Echivalentul augmentărilor din Keras
augmenter = T.Compose([
    T.RandomHorizontalFlip(),
    # tf.keras.layers.RandomRotation(0.12) reprezintă 0.12 * 360 grade = ~43 grade
    T.RandomRotation(degrees=43), 
    # tf.keras.layers.RandomZoom(0.12) -> scara între 1-0.12 și 1+0.12
    # tf.keras.layers.RandomTranslation(0.05, 0.05) -> translate max 5% pe axe
    T.RandomAffine(degrees=0, translate=(0.05, 0.05), scale=(0.88, 1.12)),
    T.ColorJitter(contrast=0.15)
])

class FundusDataset(Dataset):
    def __init__(self, data, labels, transform=None):
        self.data = torch.tensor(data, dtype=torch.float32)
        self.labels = torch.tensor(labels, dtype=torch.long)
        self.transform = transform

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        img = self.data[idx]
        label = self.labels[idx]
        
        if self.transform:
            img = self.transform(img)
            
        return img, label

# ----------------------------------------------------------
# PAPER CNN ARCHITECTURE (PyTorch)
# ----------------------------------------------------------
class PaperCNN(nn.Module):
    def __init__(self, num_classes):
        super(PaperCNN, self).__init__()
        
        self.features = nn.Sequential(
            # PyTorch padding="same" necesită stride=1
            nn.Conv2d(3, 32, kernel_size=5, padding="same"),
            nn.PReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
            
            nn.Conv2d(32, 64, kernel_size=5, padding="same"),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
            
            nn.Conv2d(64, 128, kernel_size=5, padding="same"),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
            
            nn.Conv2d(128, 256, kernel_size=5, padding="same"),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
            
            nn.AdaptiveAvgPool2d((1, 1)) # Echivalent cu GlobalAveragePooling2D
        )
        
        self.classifier = nn.Sequential(
            nn.Linear(256, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            
            nn.Linear(128, 64),
            nn.ReLU(),
            
            # Notă: Nu adăugăm Softmax aici, CrossEntropyLoss în PyTorch include Softmax.
            nn.Linear(64, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = torch.flatten(x, 1)
        x = self.classifier(x)
        return x

# ----------------------------------------------------------
# TRAIN FUNCTION
# ----------------------------------------------------------
def run_experiment(X, y_labels, num_classes, epochs, batch_size, lr, name):

    print(f"\n========== {name} ==========")

    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
    scores = []

    all_true = []
    all_pred = []

    for fold, (train_idx, val_idx) in enumerate(skf.split(X, y_labels), 1):
        print(f"\n--- Fold {fold} ---")

        X_train, X_val = X[train_idx], X[val_idx]
        y_train, y_val = y_labels[train_idx], y_labels[val_idx]

        # Datasets & DataLoaders
        train_ds = FundusDataset(X_train, y_train, transform=augmenter)
        val_ds = FundusDataset(X_val, y_val, transform=None)

        train_loader = DataLoader(train_ds, batch_size=batch_size, shuffle=True, num_workers=0)
        val_loader = DataLoader(val_ds, batch_size=batch_size, shuffle=False, num_workers=0)

        # Class weights
        weights = compute_class_weight(
            class_weight="balanced",
            classes=np.unique(y_train),
            y=y_train
        )
        class_weights = torch.tensor(weights, dtype=torch.float32).to(device)

        if name == "Severity":
            class_weights = torch.tensor([1.0, 1.4, 1.3, 2.0, 2.5], dtype=torch.float32).to(device)

        # Model Init
        model = PaperCNN(num_classes).to(device)
        
        # Multi-GPU support (echivalent MirroredStrategy)
        if gpu_count > 1:
            model = nn.DataParallel(model)

        optimizer = optim.Adam(model.parameters(), lr=lr)
        criterion = nn.CrossEntropyLoss(weight=class_weights, label_smoothing=0.1)
        
        # Scheduler (ReduceLROnPlateau)
        scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5, patience=3, min_lr=1e-7)

        # Variabile pentru Early Stopping & Checkpoint
        best_val_acc = 0.0
        patience_counter = 0
        patience_limit = 8
        best_model_path = f"{name}_fold{fold}.pth"

        # Buclele de epoci
        for epoch in range(epochs):
            model.train()
            train_loss, train_correct, train_total = 0.0, 0, 0
            
            for batch_x, batch_y in train_loader:
                batch_x, batch_y = batch_x.to(device), batch_y.to(device)
                
                optimizer.zero_grad()
                outputs = model(batch_x) # apeleaza automat model.forward(batch_x)
                loss = criterion(outputs, batch_y)
                loss.backward()
                optimizer.step()
                
                train_loss += loss.item() * batch_x.size(0) # aceasta inmultire e suma mediilor pe batch ca loss e (loss1+...)/nr img per batch
                _, predicted = torch.max(outputs, 1)
                train_total += batch_y.size(0)
                train_correct += (predicted == batch_y).sum().item()
                
            train_acc = train_correct / train_total
            train_loss = train_loss / train_total

            # Validare
            model.eval()
            val_loss, val_correct, val_total = 0.0, 0, 0
            
            with torch.no_grad():
                for batch_x, batch_y in val_loader:
                    batch_x, batch_y = batch_x.to(device), batch_y.to(device)
                    
                    outputs = model(batch_x)
                    loss = criterion(outputs, batch_y)
                    
                    val_loss += loss.item() * batch_x.size(0)
                    _, predicted = torch.max(outputs, 1)
                    val_total += batch_y.size(0)
                    val_correct += (predicted == batch_y).sum().item()

            val_acc = val_correct / val_total
            val_loss = val_loss / val_total
            
            # Print status
            current_lr = optimizer.param_groups[0]['lr']
            print(f"Epoch {epoch+1}/{epochs} - loss: {train_loss:.4f} - accuracy: {train_acc:.4f} - val_loss: {val_loss:.4f} - val_accuracy: {val_acc:.4f} - lr: {current_lr:.1e}")

            # Schedulers & Callbacks
            scheduler.step(val_acc)

            if val_acc > best_val_acc:
                best_val_acc = val_acc
                patience_counter = 0
                torch.save(model.state_dict(), best_model_path)
            else:
                patience_counter += 1

            if patience_counter >= patience_limit:
                print("Early stopping triggered!")
                break

        # După terminarea antrenamentului, încărcăm cele mai bune weight-uri (Restore Best Weights)
        model.load_state_dict(torch.load(best_model_path))
        model.eval()
        
        # Predict pe tot setul de validare pentru rapoarte și ROC
        fold_preds = []
        fold_probs = []
        fold_true = []
        
        with torch.no_grad():
            for batch_x, batch_y in val_loader:
                batch_x = batch_x.to(device)
                outputs = model(batch_x)
                probs = torch.softmax(outputs, dim=1)
                _, predicted = torch.max(outputs, 1)
                
                fold_preds.extend(predicted.cpu().numpy())
                fold_probs.extend(probs.cpu().numpy())
                fold_true.extend(batch_y.numpy())
                
        fold_probs = np.array(fold_probs)
        fold_preds = np.array(fold_preds)
        fold_true = np.array(fold_true)
        
        acc = accuracy_score(fold_true, fold_preds)
        scores.append(acc)

        all_true.extend(fold_true)
        all_pred.extend(fold_preds)

        print("Fold Accuracy:", acc)

        # Confusion Matrix
        cm = confusion_matrix(fold_true, fold_preds)
        plt.figure(figsize=(6,5))
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
        plt.title(f"{name} Fold {fold}")
        plt.show()

        # ROC for binary
        if num_classes == 2:
            auc = roc_auc_score(fold_true, fold_probs[:, 1])
            fpr, tpr, _ = roc_curve(fold_true, fold_probs[:, 1])

            plt.figure(figsize=(6,5))
            plt.plot(fpr, tpr, label=f"AUC = {auc:.3f}")
            plt.plot([0,1],[0,1],'--')
            plt.legend()
            plt.title("ROC Curve")
            plt.show()

    print("\nAverage Accuracy:", np.mean(scores))
    print("\nClassification Report:")
    print(classification_report(all_true, all_pred))

# ----------------------------------------------------------
# RUN ALL 3 EXPERIMENTS
# ----------------------------------------------------------
# 1 Binary Classification
run_experiment(
    X, y_binary,
    num_classes=2,
    epochs=30,
    batch_size=32,
    lr=1e-4,
    name="Binary"
)

# 2 Early Detection
run_experiment(
    X, y_early,
    num_classes=3,
    epochs=40,
    batch_size=16,
    lr=1e-3,
    name="Early"
)

# 3 Severity Classification
run_experiment(
    X, y_severity,
    num_classes=5,
    epochs=40,
    batch_size=8,
    lr=1e-4,
    name="Severity"
)
```
